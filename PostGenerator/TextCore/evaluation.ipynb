# Text core evaluation

#### Verificar se o tamanho é inferior a 280 carateres 

Este era um dos requisitos

```python
import pandas  as pd
import numpy as np
df = pd.read_csv('textCore.csv', index_col=0)
df.head()

df=df.replace({np.NaN:'?'})
df.head()
```

```python
total = 0  #número de linhas, deve dar 100
# Nas seguintes variaveis vamos incrementar sempre o número de carateres for inferior a 280

t5_count = 0
bert_count = 0
bart_count = 0
xlnet_count = 0
gtp2_count = 0
sumy_count = 0
nltk_count = 0
emo_count = 0
claim_count = 0

dicionary = {'T5':0, 'BERT':0, 'BART': 0, 'XLNet':0, 'GTP2':0, 'textRank_sumy':0, 'textRank_nltk':0, 'emotion_selection':0, 'claim_sellection':0}

for index, row in df.iterrows():
 
    for i in range (3,12):
        if len(row[i])>1 and len(row[i])<280: #Verifica se tem mais do 280 carateres
            
            if i==3:
                t5_count = t5_count + 1
            elif i ==4:
                bert_count = bert_count + 1
            
            elif i == 5:
                bart_count = bart_count + 1

            elif i == 6:
                xlnet_count = xlnet_count + 1
            elif i == 7:
                gtp2_count =  gtp2_count + 1
            elif i == 8:
                sumy_count = sumy_count + 1
            elif i == 9:
                nltk_count = nltk_count + 1
            elif i == 10:
                emo_count = emo_count + 1
            elif i == 11:
                claim_count = claim_count + 1
    
    total = total + 1

print(t5_count /total,bert_count/total ,
bart_count /total,
xlnet_count  /total,
gtp2_count  /total,
sumy_count  /total,
nltk_count  /total,
emo_count  /total,
claim_count  /total)

print(total)
```

Com analise previa identificou-se que o modelo bert, gtp2 e xlnet tem igual percentagem de frases com 60\% de caracteres superior a 280. 

Será que isto significa que os sumários obtidos por estes tres modelos são iguai?

```python
total_iguais = 0
for index, row in df.iterrows():
 
    if row[4] == row[6] == row[7]:
        total_iguais = total_iguais + 1

print('Da amostra de 100 noticias, o summary obtido pelos modelos bert, gtp2 e xlnet são iguais em :', total_iguais)
print('em %: ' , total_iguais/100 * 100)
```

#### Verificar fake news

```python
'''
Scrapping a ferramenta de fake-news.
https://fake-news-detection-nlp.herokuapp.com

https://github.com/ria18405/fake-news-classifier
'''

import requests
import re

def fake_news_classifier(new):

    url = "https://fake-news-detection-nlp.herokuapp.com"

    playload = {"newsTextInputForm": new,"titleInputForm": " ", "authorInputForm": " " }
    code = 0
    while (code!=200):
        res = requests.post(url, data=playload)
        code = res.status_code
        print("....")
    
    result = res.text.split('This news article is classified as a ')[1].split('</h4></div>')[0].strip()

    
    return result
```
