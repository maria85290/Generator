```python
## Fazer substituições de palavras por outras do lexico. 

f = open("NRCPortugues.txt", "r")
```

```python
lines = f.readlines()

nrc = {}
for l in lines[1:]:
    lista = l.split()
    word = lista[1]
    emo = lista[2]
    value = lista[3]

    if word not in nrc:
        nrc[word] = []

    nrc[word].append({emo:value})
```
